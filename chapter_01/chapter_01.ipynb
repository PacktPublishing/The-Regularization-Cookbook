# Overview of regularization

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
np.random.seed(42)
```

```python
def plot_data(surface: np.array, price: np.array, fit: np.array = None, legend: bool = False):
    """
    Plot the provided input data.

    Parameters
    ----------
    surface : np.array
        An array of surface values
    price : np.array
        An array of price values
    third : fit, optional
        An array of prediction values, if None will not be plotted
    legend : bool, optional
        Will display legend if True

    """
    plt.scatter(surface, price, label='data')
    if fit is not None:
        plt.plot(surface, fit, label='fit')
    if legend:
        plt.legend()
    plt.ylim(11300, 11550)
    plt.xlabel('Surface (m$^{2}$)')
    plt.ylabel('Price (â‚¬/m$^{2}$)')
    plt.grid(True)
    plt.show()
```

```python
# Define the surfaces and prices
surface = np.array([15, 17, 20, 22, 25, 28]).reshape(-1, 1)
price = 12000 - surface*50 + np.square(surface) + np.random.normal(0, 30, surface.shape)
# Plot the data
plot_data(surface, price)
```

```python
# Perform a linear regression on the data
lr = LinearRegression()
lr.fit(surface, price)
# Compute prediction
y_pred = lr.predict(surface)
# Plot data
plot_data(surface, price, y_pred, True)
```

```python
# Generate data
updated_surface = np.array([15, 17, 20, 22, 25, 28, 30, 33, 35, 37]).reshape(-1, 1)
updated_price = 12000 - updated_surface*50 + np.square(updated_surface) + np.random.normal(0, 30, updated_surface.shape)
# Plot data
plot_data(updated_surface, updated_price)
```

```python
# Perform linear regression and plot result
lr = LinearRegression()
lr.fit(updated_surface, updated_price)
y_pred_updated = lr.predict(updated_surface)
plot_data(updated_surface, updated_price, y_pred_updated, True)
```

```python
# Compute power up to 15
x_power15 = np.concatenate([np.power(updated_surface, i+1) for i in range(15)], 1)
# Perform linear regression and plot result
lr = LinearRegression()
lr.fit(x_power15, updated_price)
y_pred_power15 = lr.predict(x_power15)
plot_data(updated_surface, updated_price, y_pred_power15, True)
```

```python
# Compute power up to 2
x_power2 = np.concatenate([np.power(updated_surface, i+1) for i in range(2)], 1)
# Perform linear regression and plot result
lr = LinearRegression()
lr.fit(x_power2, updated_price)
y_pred_power2 = lr.predict(x_power2)
plot_data(updated_surface, updated_price, y_pred_power2, True)
```
